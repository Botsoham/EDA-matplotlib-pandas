```python
print("\nDataset Information:")
print(data.info())
```

```python
print("\nMissing Values:")
print(data.isnull().sum())

print("\nStatistical Summary:")
print(data.describe())
```

```python
average_salary = data['Salary'].mean()
average_experience = data['Experience'].mean()
print(f"\nAverage Salary: {average_salary:.2f}")
print(f"Average Experience: {average_experience:.2f} years")
```

```python
avg_salary_dept = data.groupby('Department')['Salary'].mean().sort_values()

plt.figure(figsize=(8,5))
avg_salary_dept.plot(kind='bar', color='skyblue')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(data['Age'], data['Salary'], color='green')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=data, x='Experience', y='Salary', hue='Department', palette='Set2', s=100)
plt.title('Experience vs Salary by Department')
plt.show()
```

```python

plt.figure(figsize=(8,6))
numeric_data = data.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
